```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina' # high res plotting

import numpy as np
import scipy as sp

import statsmodels.formula.api as smf
from sklearn import decomposition
from numpy import polyfit

import seaborn as sns
sns.set_style('whitegrid')
sns.set_style("whitegrid", {'axes.grid' : False})

import matplotlib
import matplotlib.pyplot as plt

# plot settings
font = {'family' : 'Verdana',
        'weight' : 'regular',
        'size'   : 15}
figure = {'figsize' : (20,8)}

matplotlib.rc('font', **font)
matplotlib.rc('figure', **figure)
```

```python
# make height, and make weight a function of height plus a little noise to make them highly correlated
height = np.random.uniform(0, 100, 1000)
weight = (height**2) - (2*height) + 100
weight = weight + np.random.normal(0, 1000, 1000)

height = height.T
weight = weight.T

plt.scatter(height, weight, alpha=0.5)
plt.show()
```

```python
coefficients = polyfit(height, weight, 1)
poly = np.poly1d(coefficients)
y_fit = poly(height)


plt.plot(height, weight, '.', alpha=0.3)
plt.plot(height, y_fit, '.', markersize=8.)
plt.show()
```

```python
coefficients = polyfit(height, weight, 2)
poly = np.poly1d(coefficients)
y_fit = poly(height)


plt.plot(height, weight, '.', alpha=0.3)
plt.plot(height, y_fit, '.', markersize=8.)
plt.show()
```

```python
from scipy.interpolate import interp1d

f2 = interp1d(height, weight, kind='zero', fill_value="extrapolate")
xnew = np.linspace(0, 100, num=1000, endpoint=True)

plt.plot(height, weight, '.', alpha=0.3)
plt.plot(xnew, f2(xnew), '-', markersize=8.)
plt.show()
plt.show()
```
